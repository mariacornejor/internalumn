```python
from keras.models import Sequential
from keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
```

```python
data = pd.read_csv('data.csv')
```

```python
features = list(data.columns)
```

```python
features
```

```python
features = data.drop(['id', 'diagnosis'], axis=1)
```

```python
label_encoder = LabelEncoder()
labels = label_encoder.fit_transform(data['diagnosis'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, labels, test_size=0.2, random_state=42)
```

```python
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(30,)))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(X_train, y_train, epochs=10, batch_size=32)
```

```python
loss, accuracy = model.evaluate(X_test, y_test)
print('Loss:', loss)
print('Accuracy:', accuracy)
```

```python
columnas = list[data.columns]
```

```python
# ID específico a predecir
id_especifico = 1   # ID del tumor específico a predecir

# Obtener las características del ID específico
datos_id = data.loc[data['id'] == id_especifico, columnas]

# Normalizar los datos del ID específico si es necesario
datos_id = scaler.transform(datos_id)

# Realizar la predicción
prediccion = model.predict(datos_id)

# Interpretar la predicción
umbral = 0.5
if prediccion >= umbral:
    resultado = "maligno"
else:
    resultado = "benigno"

print(f"El tumor con ID {id_especifico} se predice como: {resultado}")
```
